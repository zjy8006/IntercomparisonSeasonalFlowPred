```python
import pandas as pd

station_names = [ # Tangnaihai
    '玛多',# 1957-2019
    '达日',# 1956-2019
    # '甘德',
    '久治',# 1958-2019
    '红原',#1960-2019
    '若尔盖',#1959-2019
    '玛曲',#1969-2019
    '玛沁',#1959-2019
    '河南',#1959-2019
    # '泽库',
    # '同德',
    '兴海',#1960-2019
    '贵南',#1960-2019
    '共和',#1955-2019
    '贵德',#1959-2019
    '同仁',#1960-2019
]
```

```python
precp = pd.DataFrame(index=pd.date_range(start='1969-01-01', end='2019-12-01', freq='MS'),columns=station_names)
max_temp = pd.DataFrame(index=pd.date_range(start='1969-01-01', end='2019-12-01', freq='MS'),columns=station_names)
min_temp = pd.DataFrame(index=pd.date_range(start='1969-01-01', end='2019-12-01', freq='MS'),columns=station_names)

precp.index.name = 'DATE'
max_temp.index.name = 'DATE'
min_temp.index.name = 'DATE'

for station in station_names:
    df = pd.read_csv(f'D:/DataSpace/HydroMeteAnthropicDatabase/7.FilledRawMeteObsInfo/ChinaLandDailyMeteV3(InsertSolarRadiation)/{station}.csv',index_col=['DATE'], parse_dates=['DATE'])
    df_p2020 = df.loc['1969-01-01':'2019-12-01', 'P2020(mm)']
    precp[station] = df_p2020.resample('MS').sum()

    df_tmax = df.loc['1969-01-01':'2019-12-01', 'MAX-TEM(C)']
    max_temp[station] = df_tmax.resample('MS').mean()

    df_tmin = df.loc['1969-01-01':'2019-12-01', 'MIN-TEM(C)']
    min_temp[station] = df_tmin.resample('MS').mean()

precp.to_csv('../data/upper_reach_yellow_river_pcp.csv')
max_temp.to_csv('../data/upper_reach_yellow_river_tmax.csv')
min_temp.to_csv('../data/upper_reach_yellow_river_tmin.csv')

precp
```

```python
hydro_mete_stations = {
    "Tangnaihai": [
        '玛多', '达日', '久治', '红原', '若尔盖', '玛曲', '玛沁', '河南', '兴海'
    ],
    "Guide": [
        '玛多', '达日', '久治', '红原', '若尔盖', '玛曲', '玛沁', '河南', '兴海',
        '贵南', '共和', '贵德', '同仁'
    ],
    "Xunhua": [
        '玛多', '达日', '久治', '红原', '若尔盖', '玛曲', '玛沁', '河南', '兴海',
        '贵南', '共和', '贵德', '同仁'
    ]
}

station_area = pd.read_csv('../data/MeteGaugeStationControlArea.csv')
station_area_dict = dict(zip(station_area['station'], station_area['Shape_Area']))

# Compute ratios for each location
ratios = {}
for location, stations in hydro_mete_stations.items():
    total_area = sum(station_area_dict[s] for s in stations)
    ratios[location] = {s: station_area_dict[s] / total_area for s in stations}

# Display the ratios
for location, station_ratios in ratios.items():
    print(f"\n{location} ratios:")
    for station, ratio in station_ratios.items():
        print(f"  {station}: {ratio:.4f}")
tangnaihai_mete_df = pd.DataFrame(index=precp.index,columns=['precp','max_temp','min_temp'])
guide_mete_df = pd.DataFrame(index=precp.index,columns=['precp','max_temp','min_temp'])
xunhua_mete_df = pd.DataFrame(index=precp.index,columns=['precp','max_temp','min_temp'])

tangnaihai_mete_df.index.name='date'
guide_mete_df.index.name='date'
xunhua_mete_df.index.name='date'

for date in tangnaihai_mete_df.index:
    precp_ = 0.0
    max_temp_ = 0.0
    min_temp_ = 0.0
    for mete_station_ration_item in ratios['Tangnaihai'].items():
        mete_stat = mete_station_ration_item[0]
        mete_stat_ratio = mete_station_ration_item[1]
        precp_ += precp.loc[date,mete_stat]*mete_stat_ratio
        max_temp_ += max_temp.loc[date,mete_stat]*mete_stat_ratio
        min_temp_ += min_temp.loc[date,mete_stat]*mete_stat_ratio
    tangnaihai_mete_df.loc[date,'precp'] = precp_
    tangnaihai_mete_df.loc[date,'max_temp'] = max_temp_
    tangnaihai_mete_df.loc[date,'min_temp'] = min_temp_

for date in guide_mete_df.index:
    precp_ = 0.0
    max_temp_ = 0.0
    min_temp_ = 0.0
    for mete_station_ration_item in ratios['Guide'].items():
        mete_stat = mete_station_ration_item[0]
        mete_stat_ratio = mete_station_ration_item[1]
        precp_ += precp.loc[date,mete_stat]*mete_stat_ratio
        max_temp_ += max_temp.loc[date,mete_stat]*mete_stat_ratio
        min_temp_ += min_temp.loc[date,mete_stat]*mete_stat_ratio
    guide_mete_df.loc[date,'precp'] = precp_
    guide_mete_df.loc[date,'max_temp'] = max_temp_
    guide_mete_df.loc[date,'min_temp'] = min_temp_

for date in xunhua_mete_df.index:
    precp_ = 0.0
    max_temp_ = 0.0
    min_temp_ = 0.0
    for mete_station_ration_item in ratios['Xunhua'].items():
        mete_stat = mete_station_ration_item[0]
        mete_stat_ratio = mete_station_ration_item[1]
        precp_ += precp.loc[date,mete_stat]*mete_stat_ratio
        max_temp_ += max_temp.loc[date,mete_stat]*mete_stat_ratio
        min_temp_ += min_temp.loc[date,mete_stat]*mete_stat_ratio
    xunhua_mete_df.loc[date,'precp'] = precp_
    xunhua_mete_df.loc[date,'max_temp'] = max_temp_
    xunhua_mete_df.loc[date,'min_temp'] = min_temp_


tangnaihai_mete_df.to_csv('../data/Tangnaihai_catchment_mete.csv')
guide_mete_df.to_csv('../data/Guide_catchment_mete.csv')
xunhua_mete_df.to_csv('../data/Xunhua_catchment_mete.csv')
    



```
