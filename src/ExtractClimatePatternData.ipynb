```python
station = pd.read_excel('../data/fork.xls',sheet_name='Sheet1')

# parse ID and NAME as a dictionary
station_dict = dict(zip(station['ID'], station['NAME']))
print(station_dict)
# parse the data
start_date = '2015-01-01'
end_date = '2019-12-31'
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
scenarios = [
    'ssp126',
    'ssp245',
    'ssp585'
]
variables = [
    'hurs', # Near-surface relative humidity
    'pr', # Precipitation
    'rsds', # Surface downwelling shortwave radiation
    'sfcWind',# Near-surface wind speed
    'tasmax', # Daily maximum near-surface air temperature
    'tasmin', # Daily minimum near-surface air temperature
]
columns_dict= {'hurs':'RHU(%)','pr':'P(mm)','rsds':'SR(MJ/m^2)','sfcWind':'WS(m/s)','tasmax':'Tmax(C)','tasmin':'Tmin(C)'}
station_IDs = ['001','002','003','004','005','006','007','008','009','010','011','012','013']
for scenario in scenarios:
    for station_ID in station_IDs:
        df = pd.DataFrame(index=date_range)
        df.index.name = 'date'
        save_path = "../data/UpperYellowRiver_ACCESS-ESM1-5/{}/".format(scenario)
        if not os.path.exists(save_path):
            os.makedirs(save_path)
        print(save_path)
        for variable in variables:
            # print(scenario,variable,scenario,station_ID)
            # read the data
            data_path = "../data/UpperYellowRiver_ACCESS-ESM1-5/{}_day_ACCESS-ESM1-5_{}_{}".format(variable,scenario,station_ID)
            file_path = data_path+".txt"
            print(file_path)
            if not os.path.exists(file_path):
                print("File not found: {}".format(file_path))
                continue
            data = pd.read_table(file_path)
            print(data["20150101"])
            df[columns_dict[variable]] = data["20150101"].values
        print(df)
        df.to_csv(save_path+'{}.csv'.format(station_dict[int(station_ID)]))
```
